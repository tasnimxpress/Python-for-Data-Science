```python
# another approach
df = pd.read_csv(dataurl, index_col=([0]))
df.head(5)
```

```python
# sample method
# default is random, by providing random_state it become static
df.sample(random_state=12)
```

```python
# n parameter 
# specify number of records we want to get back
df.sample(n=3)
```

```python
# frac parameter
# percentage of data we want to get back
df.sample(frac=0.01)
```

```python
# sampling with replacement
df.sample(n=3, replace=True)
```
