```python
# data source
dataurl = 'https://andybek.com/pandas-nutrition'
```

```python
# import pandas
import pandas as pd
```

```python
# read data
df = pd.read_csv(dataurl)
df.head(5)
```

```python
# info
df.info(verbose=False, memory_usage='deep')
```

```python
# remove unnamed column - duplicate of index
df['Unnamed: 0']
```

```python
df.drop('Unnamed: 0', axis=1)
```

```python
df.head(5)
```

```python
# another approach
df = pd.read_csv(dataurl, index_col=([0]))
df.head(5)
```

```python
# sample method
# default is random, by providing random_state it become static
df.sample(random_state=12)
```

```python
# n parameter 
# specify number of records we want to get back
df.sample(n=3)
```

```python
# frac parameter
# percentage of data we want to get back
df.sample(frac=0.01)
```

```python
# sampling with replacement
df.sample(n=3, replace=True)
```
